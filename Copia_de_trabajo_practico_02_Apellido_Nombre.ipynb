## Punto 2

Mostrar por pantalla los nombres (columnas Name) ordenados alfabéticamente y segmentado por clase (columna Pclass) y numerados desde 0 para cada clase.

Ejemplo:

Clase x

0 nombre

1 nombre

...


Clase y

0 nombre

1 nombre

...

```python
# Ordenar los nombres alfabéticamente y segmentar por clase
for clase in sorted(df['Pclass'].unique()):
    print(f'\nClase {clase}\n')
    # Filtrar los pasajeros de cada clase
    clase_df = df[df['Pclass'] == clase].sort_values(by='Name')
    # Numerar los nombres desde 0
    for idx, name in enumerate(clase_df['Name']):
        print(f'{idx} {name}')
```


## Punto 3

Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron  (columna Survived).

El valor 1 indica que está sobrevivio y 0 no.


```python
# Contar la cantidad total de pasajeros
total_passengers = len(df)

# Contar cuántos sobrevivieron (Survived == 1) y cuántos murieron (Survived == 0)
survived_count = df['Survived'].sum()  # Suma de 1's, los que sobrevivieron
died_count = total_passengers - survived_count  # El resto, los que murieron

# Calcular los porcentajes
survived_percentage = (survived_count / total_passengers) * 100
died_percentage = (died_count / total_passengers) * 100

# Mostrar los resultados
print(f'Porcentaje de personas que sobrevivieron: {survived_percentage:.2f}%')
print(f'Porcentaje de personas que murieron: {died_percentage:.2f}%')
```

## Punto 4

Mostrar por pantalla el porcentaje de personas que sobrevivieron (columna Survived) en cada clase (columna Pclass).

El valor 1 indica que sobrevivio y 0 no.

```python
# Calcular el porcentaje de sobrevivientes por clase
survived_by_class = df.groupby('Pclass')['Survived'].mean() * 100

# Mostrar los resultados
for clase, porcentaje in survived_by_class.items():
    print(f'Porcentaje de personas que sobrevivieron en Clase {clase}: {porcentaje:.2f}%')
```

## Punto 5

Realizar un gráfico de barras horizontal con la cantidad de personas según la edad (columna Age). La franja es de 10 años, es decir, 0-9, 10-19, ..., 90-99.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que ya tienes el DataFrame 'df' cargado con el dataset

# Crear las franjas de edad
bins = range(0, 101, 10)  # Franjas de 0 a 100 años en intervalos de 10
labels = [f"{i}-{i+9}" for i in bins[:-1]]  # Crear etiquetas como '0-9', '10-19', etc.

# Segmentar la edad en las franjas
df['Age Group'] = pd.cut(df['Age'], bins=bins, right=False, labels=labels)

# Contar la cantidad de personas en cada grupo de edad
age_counts = df['Age Group'].value_counts().sort_index()

# Crear el gráfico de barras horizontal
plt.figure(figsize=(10, 6))
age_counts.plot(kind='barh', color='skyblue')
plt.title('Cantidad de Personas por Grupo de Edad')
plt.xlabel('Cantidad de Personas')
plt.ylabel('Grupo de Edad')
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--')
plt.show()
```
